```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
movie = pd.read_csv('tmdb_5000_movies.csv')
credit = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movie.head(1)
```

```python
credit.head(1)
```

```python
movie = movie.merge(credit,on='title')
```

```python
movie = movie[['id','title','overview','genres','keywords','cast','crew']]
movie.head(1)
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace=True)
```

```python
movie.isnull().sum()
```

```python
movie.duplicated().sum()
```

```python
import ast
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):               # this method for converting string data to interger
        l.append(i['name'])
    return l

movie['genres'] = movie['genres'].apply(convert)
movie.head(5)
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
```

```python
def convert1(obj):
    l = []
    count = 0
    for i in ast.literal_eval(obj):
        if count !=3:
            l.append(i['name'])
            count += 1
        else:
            break
    return l

movie['cast'] = movie['cast'].apply(convert1)
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
movie['crew'] = movie['crew'].apply(fetch_director)
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie.head(5)
```

```python

col = ['genres','keywords','cast','crew']

for i in col:
    movie[i] = movie[i].apply(lambda x:[i.replace(" ","")for i in x])



'''
movie['genres'] = movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['keywords'] = movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast'] = movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew'] = movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
'''
movie.head()
```

```python
movie['tags'] = movie['overview'] + movie['genres'] + movie['keywords'] + movie['cast'] + movie['crew']
movie.head()
```

```python
df = movie[['id','title','tags']]
```

```python
df['tags'] = df['tags'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movies):
    movie_index = df[df['title'] == movies].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
